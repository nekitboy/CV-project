```

MODEL_CFG         = 'cfg/yolo_custom.cfg'
MODEL_BATCH_SIZE  = 8
MODEL_EPOCH       = 15
MODEL_LR          = 0.00001
YOLO_WEIGHTS_PATH = 'yolo.weights'
DATASET_PATH      = 'dataset/train/'
XML_PATH = 'yolo_xml/'
CONFIDENCE_LEVEL  = 0.06
clear_output()
```

```
options = {
    "model"     : MODEL_CFG, 
    "load"      : YOLO_WEIGHTS_PATH,
    "batch"     : MODEL_BATCH_SIZE,
    "epoch"     : MODEL_EPOCH,
    "lr"        : MODEL_LR,
    "annotation": XML_PATH,
    "dataset"   : DATASET_PATH,
    "gpu"       : 1.0,
    "train"     : True
}
# initialize and train model
tfnet = TFNet(options)
tfnet.train()
# save weights
tfnet.savepb()
```
